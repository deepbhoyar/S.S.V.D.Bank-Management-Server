```python
import mysql.connector as sql
mydb = sql.connect(host='localhost',user='root',password='deep123',database='BANK_MANAGEMENT')
print(mydb)
x = mydb.cursor()
```

```python
class bank_management:
    
    def OpenAcc(self):
        n = input("Enter The Name: ")
        ac = input("Enter The Account No: ")
        db = input("Enter The Date Of Birth : ")
        add = input("Enter The Address : ")
        cn = input("Enter The Contact Number : ")
        ob = int(input("Enter The Opening Balance: "))
        data1 = (n,ac,db,add,cn,ob)
        data2 = (n,ac,ob)
        sql1 = ('insert into Account values (%s,%s,%s,%s,%s,%s)')
        sql2 = ('insert into Amount values (%s,%s,%s) ')
        x = mydb.cursor()
        x.execute(sql1,data1)
        x.execute(sql2,data2)
        mydb.commit()
        print("Data Entered and account opened Successfully,WELCOME TO S.S.V.D FAMILY...")
        
    

    def depoAmo(self):
        amount = input("Enter the amount you want to deposit: ")
        ac = input("Enter The Account No: ")
        a = 'select balance from Amount table where AccNo=%S'
        data = (ac,) #tuple creation
        x = mydb.cursor()
        x.execute(a,data)
        result = x.fetchone()
        t = result[0] + amount
        sql = ('update Amount set balance where AccNo=%s')
        d = (t,ac)
        x.execute(sql,d)
        mydb.commit()
       
        

    def withdrawAmount(self):
        amount = input("Enter the amount you want to withdraw: ")
        ac = input("Enter The Account No: ")
        a = 'select balance from amount table where AccNo=%S'
        data = (ac,)
        x = mydb.cursor()
        x.execute(a,data)
        result = x.fetchone()
        t = result[0] - amount
        sql = ('update amount set balance where AccNo=%s')
        d = (t,ac)
        x.execute(sql,d)
        mydb.commit()
        print("Amount withdrawl successfull")
        

    def balEnq(self):
        ac = input("Enter The Account No: ")
        a ='select * from Amount where AccNo=%s'
        data = (ac,)
        x= mydb.cursor()
        x.execute(a,data)
        result=x.fetchone()
        print("Balance for the account:",ac,"is",result[-1])
       

    def disDetails(self):
        ac = input("Enter The Account No: ")
        a ='select * from account where AccNo =%s'
        data = (ac,)
        x = mydb.cursor()
        x.execute(a,data)
        result=x.fetchone()
        for i in result:
            print(i)
        

    def closeAcc(self):
        ac = input("Enter The Account No: ")
        sql1 = 'delete from Account where AccNo = %s'
        sql2 = 'delete from Amount where AccNo = %s'
        data = (ac,)
        x = mydb.cursor()
        x.execute(sql1,data)
        x.execute(sql2,data)
        mydb.commit()
        print('Your account has been closed, THANKYOU FOR BANKING WITH US. !!!')
        
    
while True:
    print("++++++++++++++++++++++++++++")
    print("**WELCOME TO S.S.V.D.BANK***")
    print("++++++++++++++++++++++++++++")
          
    print("please select your services")
          
    print('''
              1.OPEN NEW ACCOUNT
              2.DEPOSIT AMOUNT  
              3.WITHDRAW AMOUNT
              4.BALANCE ENQUIRY
              5.DISPLAY CUSTOMER DETAILS
              6.CLOSE AN ACCOUNT''')

    choice  = input("Enter The Task You Want To Perform : ")
    s=bank_management()
    if choice=='1':
        s.OpenAcc()
    elif choice=='2':
        s.depoAmo()
    elif choice=='3':
        s.withdrawAmount()
    elif choice=='4':
        s.balEnq()
    elif choice=='5':
        s.disDetails()
    elif choice=='6':
        s.closeAcc()
    else:
        print("InvalidÂ Choice")
        
```
